Goal: Manipulate real images using NumPy + OpenCV.

### 1. Import Required Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### 2. Load Image

```python
image_path = "../data/images/test.png"
image = cv2.imread(image_path)
print("Shape:", image.shape)
```

### 3. Display Image Properly

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

### 4. Resize Image

```python
#         cv2.resize(image, (width, height))
resized = cv2.resize(image, (300, 300))
resized_rgb = cv2.cvtColor(resized, cv2.COLOR_BGR2RGB)
plt.imshow(resized_rgb)
plt.axis("off")
plt.show()
```

### 5. Crop Image (Slicing)

```python
#         image[row_start:row_end, col_start:col_end]
cropped = image[100:400, 200:500]
cropped_rgb = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)
plt.imshow(cropped_rgb)
plt.axis("off")
plt.show()
```

### 6. Draw Rectangle (Bounding Box)

```python
image_copy = image.copy()
# cv2.rectangle(image, start_point, end_point, color, thickness)
cv2.rectangle(image_copy, (200, 100), (500, 400), (0, 255, 0), 3)
image_copy_rgb = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy_rgb)
plt.axis("off")
plt.show()
```

### 7. Edge Detection

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# cv2.Canny(image, threshold1, threshold2)
edges = cv2.Canny(gray, 100, 200)
plt.imshow(edges, cmap="gray")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
edges_low = cv2.Canny(gray, 1, 10)
edges_high = cv2.Canny(gray, 200, 250)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Low Threshold")
plt.imshow(edges_low, cmap="gray")
plt.axis("off")
plt.subplot(1,2,2)
plt.title("High Threshold")
plt.imshow(edges_high, cmap="gray")
plt.axis("off")
plt.show()
```

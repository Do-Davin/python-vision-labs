### 1. Import Required Libraries

```python
import cv2 # image handling
import numpy as np # arrays
import matplotlib.pyplot as plt # display image
```

### 2. Load Image from file

```python
# Read image into Numpy array.
image_path = "../data/images/test.png"
image = cv2.imread(image_path)
print(type(image))
```

### 3. Inspect Image Type and Shape

```python
# Understand image dimensions and structure.
print("Image shape:", image.shape)
```

### 4. Display Image Correctly (BGR â†’ RGB)

```python
# Show image properly using matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

### 5. Access a Single Pixel

```python
# Understand how to read pixel values.
# Choose pixel position (row, column)
row = 100
col = 200
pixel_value = image[row, col]
print("Pixel value at (100, 200):", pixel_value)
```

### 6. Modify a Pixel

```python
# Learn how to change pixel values.
# Change pixel at (100, 200) to pure red (BGR format)
image[100:110, 200:210] = [0, 0, 255]
# Convert to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis("off")
plt.show()
```

### 7. Convert Image to Grayscale

```python
# Understand channel reduction (3 -> 1)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap="gray")
plt.axis("off")
plt.show()
print("Original shape:", image.shape)
print("Grayscale shape:", gray_image.shape)
```
